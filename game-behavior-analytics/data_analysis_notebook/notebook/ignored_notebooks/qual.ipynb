```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Read data
df = pd.read_csv('../data/qualitative_analysis_codes.csv')

# Rename the unnamed first column
df = df.rename(columns={df.columns[0]: 'code'})

# Convert percentage strings to numeric for concept columns
concept_cols = [col for col in df.columns if col not in ['code', 'cheater_type']]
for col in concept_cols:
    df[col] = df[col].str.rstrip('%').astype(float)


# Print summary table
print("Summary: Average % across all concepts for each code-cheater type:")
summary = df.groupby(['code', 'cheater_type'])[concept_cols].mean().mean(axis=1).unstack()
print(summary.round(1))
```

```python
# Read data
df = pd.read_csv('../data/qualitative_analysis_codes.csv')

# Rename the unnamed first column
df = df.rename(columns={df.columns[0]: 'code'})

# Convert percentage strings to numeric for concept columns
concept_cols = [col for col in df.columns if col not in ['code', 'cheater_type']]
for col in concept_cols:
    df[col] = df[col].str.rstrip('%').astype(float)

# Create a comprehensive table showing all 15 concepts
print("Table: Percentage values for all 15 concepts by code and cheater type")
print("="*80)

# Pivot the data to show all concepts
for concept in concept_cols:
    print(f"\n{concept.upper()}:")
    concept_table = df.pivot(index='code', columns='cheater_type', values=concept)
    print(concept_table.round(1))
    print("-" * 50)
```
